```python
msk = np.random.randn(len(df_cars)) < 0.8
train = df_cars[msk]
test  = df_cars[~msk]
print(msk)
print(~msk)
print(train)
print(test)
```

```python
plt.scatter(train.Speed , train.CO2 , c = 'r')
plt.xlabel('speed')
plt.ylabel('CO2')
plt.show()
```

```python
plt.scatter(test.Speed , test.CO2 , c = 'b')
plt.xlabel('speed')
plt.ylabel('CO2')
plt.show()
```

```python
fig = plt.figure()
axe = fig.add_subplot(111)
axe.scatter(train.Speed , train.CO2 , c = 'r')
axe.scatter(test.Speed , test.CO2 , c = 'b')
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x =  np .asanyarray(train['Speed']).reshape(-1,1)
train_y = np .asanyarray(train['CO2'])
regr.fit(train_x , train_y)
print( 'Coef_ :'   ,regr.coef_)
print( 'intercept :' , regr.intercept_)
```

```python
plt.scatter(train.Speed ,  train.CO2 , c="red")
plt.plot(train_x,regr.coef_*train_x + regr.intercept_ , '--g')
plt.show()
```

```python
from sklearn.metrics import r2_score
test_x = np.asanyarray(test['Speed']).reshape(-1,1)
test_y = np.asanyarray(test['CO2'])
test_y_ = regr.predict(test_x)
print('r2_Score  :' , r2_score(test_y , test_y_))
```
